Установим библиотеку:

```r
install.packages("DescTools")
```

Запустим тест:

```r
library(DescTools)

data_table <- matrix(c(13, 19, 19, 25,   # Строка правильных ответов
                       63, 57, 57, 51),  # Строка неправильных ответов
                     nrow = 2,
                     byrow = TRUE)

colnames(data_table) <- c("Difficult_1", "Difficult_2", "Difficult_3", "Difficult_4")
rownames(data_table) <- c("True", "False")

print(data_table)

CochranArmitageTest(data_table, alternative = "one.sided")
```

```r
citation("DescTools")
```
